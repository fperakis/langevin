```python
from langevin.simulation import run_langevin_simulation
from langevin.visualization import compute_local_diffusion
from langevin.visualization import plot_trajectory_colored_by_diffusion
```

```python
# simulations parameters
dt = 1e-2
n_particles = 2
t_max = 1e3
n_steps = int(t_max/dt)
m=1.0
kB=1.
T = 0.0115#1
gamma_l = 1.
gamma_g = 1e2
tau = 1
```

```python
# run simulation
positions, velocities, gamma_values = run_langevin_simulation(
    T=T,tau=tau,n_steps=n_steps,n_particles=n_particles,dt=dt,gamma_l=gamma_l, gamma_g=gamma_g,m=m,kB=kB)
```

```python
i_particle = 0
D_local, mid_points  = compute_local_diffusion(positions[i_particle], dt=0.01, window_size=50)
plot_trajectory_colored_by_diffusion(positions[i_particle,1:], D_local, mid_points, vmin=0)
```
